```python
import os
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
from matplotlib import gridspec
import os
import numpy as np
import scipy.signal
import librosa
import librosa.display
import matplotlib.pyplot as plt
from matplotlib import gridspec
import gspread
from google.colab import drive
drive.mount('/content/gdrive')
from google.colab import auth
auth.authenticate_user()
from google.colab import files
import gspread
from google.auth import default
creds, _ = default()
import pandas as pd

import os
import numpy as np
import librosa
import librosa.display
import pandas as pd

# Function to Extract Features
def extract_features(y, sr):
    chroma_stft = np.mean(librosa.feature.chroma_stft(y=y, sr=sr))
    spectral_centroid = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr))
    spectral_bandwidth = np.mean(librosa.feature.spectral_bandwidth(y=y, sr=sr))
    spectral_rolloff = np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr))
    rmse = np.mean(librosa.feature.rms(y=y))
    zero_crossing_rate = np.mean(librosa.feature.zero_crossing_rate(y))

    return [chroma_stft, spectral_centroid, spectral_bandwidth, spectral_rolloff, rmse, zero_crossing_rate]

# Create an empty DataFrame to store results
results = {"Filename": [], "Duration (seconds)": [], "Chroma_stft": [], "Spectral_centroid": [],
           "Spectral_bandwidth": [], "Spectral_rolloff": [], "Rmse": [], "Zero_crossing_rate": []}

# Folder containing MP3 files
folder_path = "/content/gdrive/MyDrive/language_music-dataset-main/Speech_Slovak"

# Iterate through the MP3 files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".mp3"):
        mp3_path = os.path.join(folder_path, filename)

        # Load the audio file
        y, sr = librosa.load(mp3_path)

        # Calculate duration
        duration = librosa.get_duration(y=y, sr=sr)

        # Extract features
        features = extract_features(y, sr)

        # Add the results to the DataFrame
        results["Filename"].append(filename)
        results["Duration (seconds)"].append(duration)
        results["Chroma_stft"].append(features[0])
        results["Spectral_centroid"].append(features[1])
        results["Spectral_bandwidth"].append(features[2])
        results["Spectral_rolloff"].append(features[3])
        results["Rmse"].append(features[4])
        results["Zero_crossing_rate"].append(features[5])

# Create a DataFrame from the results dictionary
df = pd.DataFrame(results)

# Save the DataFrame to a CSV file
csv_file_path = "/content/gdrive/MyDrive/language_music-dataset-main/Speech_Slovak_dataset.csv"
df.to_csv(csv_file_path, index=False)

# Print the results DataFrame
print(df)
```
